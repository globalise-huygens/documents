```python
import sqlite3

DB_PATH = "globalise_documents.db"


def get_document_ids_for_scan(filename):
    """
    Return a list of distinct document IDs associated with a scan filename.
    """
    query = """
    SELECT DISTINCT d.id
    FROM scan s
    JOIN page p
        ON p.scan_id = s.id
    JOIN page2document p2d
        ON p2d.page_id = p.id
    JOIN document d
        ON d.id = p2d.document_id
    WHERE s.filename = ?;
    """

    with sqlite3.connect(DB_PATH) as conn:
        cursor = conn.cursor()
        cursor.execute(query, (filename,))
        return [row[0] for row in cursor.fetchall()]
```

```python
get_document_ids_for_scan('NL-HaNA_1.04.02_8793_0024')
```
